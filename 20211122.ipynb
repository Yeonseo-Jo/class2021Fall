```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

A  = 440.00
C  = 523.25
F  = 698.46 
D  = 587.33
E  = 659.25	 
Eb = 622.25 
F  = 698.46 
G  = 783.99
Bb = 932.33
```

```python
#떴다 떴다 비행기

notes = [E,D,C,D,E,E,E,D,D,D,E,E,E,E,D,C,D,E,E,E,D,D,E,D,C]

durs = [0.75, 0.25, 0.5, 0.5, 0.5,0.5, 1.0, 0.5, 0.5, 1.0, 0.5, 0.5, 1.0, 0.75, 0.25, 0.5, 0.5, 0.5, 0.5, 1.0, 0.5,0.5,0.5,0.5,2]
```

```python
s_all = np.empty(0)   

for note, dur in zip(notes, durs):        
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```
