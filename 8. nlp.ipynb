### Tokenization

= 단어별로 쪼개주는 기능

```python
import nltk
```

```python
text = 'Here’s to the crazy ones, the misfits, the rebels, the troublemakers, the round pegs in the square holes. \
The ones who see things differently — they’re not fond of rules. \
You can quote them, disagree with them, glorify or vilify them, \
but the only thing you can’t do is ignore them because they change things. \
They push the human race forward, and while some may see them as the crazy ones, we see genius, \
because the ones who are crazy enough to think that they can change the world, are the ones who do.'
```

```python
text.split()
```

```python
' '.join(text.split())
```

```python
#split도 tokenize 기능을 하지만, word_tokenize가 더 advanced 된 기능을 수행해줌

from nltk.tokenize import word_tokenize
nltk.download('punkt')      #다운로드 반드시 해줘야 함(word_tokenize에서는 word 뿐만 아니라 punktation도 tokenize하기 때문. 그래서 punkt에 대한 데이터가 필요한 것)
#구두점도 같이 tokenize됨
words = word_tokenize(text)
words
```

```python
#regular expression을 기반으로 한 tokenize

from nltk.tokenize import RegexpTokenizer
retokenize = RegexpTokenizer("[\w]+")       #\w+= 알파벳 혹은 숫자 문자의 연쇄 = words의 연쇄 추출
words = retokenize.tokenize(text)
words
```

### Normalization  
Stemming: am → am, the going → the go, having → hav  
= 원형을 찾아줌 (단어 원형 부분을)

Lemmatization: am → be, the going → the going, having → have
= 원형 찾아줌 (사전에 등장한 진짜 단어 원형을 찾아줌)

```python
#stemming - (1)

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
[stemmer.stem(w) for w in words]     #words에 있는 단어들 stemming
```

```python
#stemming-(2)

from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
[stemmer.stem(w) for w in words]
```

```python
#lemmatization

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
[lemmatizer.lemmatize(w) for w in words]
```

### Stopword

=조사, 대명사,전치사 등 너무 빈번하게 나오는 단어는 제외하는 기능

```python
from nltk.corpus import stopwords  
nltk.download('stopwords')
print(words)
words = [w for w in words if not w in stopwords.words('english')] #stopwords 단어들 제외 단어 출력
print(words)
```

### Collocation, Concordance

* collocation : 빈번하게 나오는 단어 쌍들을 찾아줌

* concordance : 예문, 용례 찾아줌

```python
nltk.download('gutenberg')  #nltk에 있는 gutenberg (=저작권이 없는 txt들 모음) 다운
text = nltk.corpus.gutenberg.raw('austen-emma.txt') #gutneberg 속 austen-emma 텍스트 다운
words = retokenize.tokenize(text) #단어별로 분리
```

```python
nltk.Text(words).collocations() #빈번히 나오는 두 단어 쌍들 찾기
```

```python
nltk.Text(words).concordance('Emma', 79, 10) #emma의 용례,용법을 소설 속에서 찾기 (찾을 논항, 왼쪽부터 논항 포함하여 찾을 단어 개수, 오른쪽부터 논항 포함하여 찾을 단어 개수)
```

```python
nltk.Text(words).dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"])
#0부터 20만까지 해당 단어가 어디에서 얼마나 나오는지를 차트로 보여줌
```

```python
nltk.Text(words).similar("Emma") #similar: Emma와 비슷하게 쓰이는 단어들을 찾앙줌
```

```python
nltk.Text(words).common_contexts(["Emma", "she"]) #emma와 she가 같이 나오는 context들을 찾아줌
```

### Regular expression

= 정규표현

```python
import re
```

```python
'''       Basic Regular Expression Meta-Characters, Including Wildcards, Ranges and Closures

.	        Wildcard, matches any character
^abc	    Matches some pattern abc at the start of a string
abc$	    Matches some pattern abc at the end of a string
[abc]	    Matches one of a set of characters
[^abc]    Matches anything but a set of characters
[A-Z0-9]	Matches one of a range of characters
ed|ing|s	Matches one of the specified strings (disjunction)
*	        Zero or more of previous item, e.g. a*, [a-z]* (also known as Kleene Closure)
+	        One or more of previous item, e.g. a+, [a-z]+
?	        Zero or one of the previous item (i.e. optional), e.g. a?, [a-z]?
{n}	      Exactly n repeats where n is a non-negative integer
{n,}	    At least n repeats
{,n}	    No more than n repeats
{m,n}	    At least m and no more than n repeats
a(b|c)+	  Parentheses that indicate the scope of the operators
(...)     Matches whatever regular expression is inside the parentheses
\d
Matches any decimal digit; this is equivalent to the class [0-9].
\D
Matches any non-digit character; this is equivalent to the class [^0-9].
\s
Matches any whitespace character; this is equivalent to the class [ \t\n\r\f\v].
\S
Matches any non-whitespace character; this is equivalent to the class [^ \t\n\r\f\v].
\w
Matches any alphanumeric character; this is equivalent to the class [a-zA-Z0-9_].
\W
Matches any non-alphanumeric character; this is equivalent to the class [^a-zA-Z0-9_].

'''

#정규표현은 대부분 string에 대해 적용
#(리스트에 regular expression을 적용하고 싶으면 loop를 돌려 string의 형태로 만들어줘야 함)
```

```python
engdict = nltk.corpus.words.words('en')

#'ed'로 끝나는 말이 있으면(re.search()가 true이면), 그 단어 통으로 리스트를 다시 만들어라 -> 루프를 돌면 ed로 끝나는 말들만 리스트에 담기게 됨
result = [w for w in engdict if re.search('ed$', w)]
# result = [w for w in engdict if re.search('^..j..t..$', w)] #j가 3번째 t가 6번째인 말들을 찾아라
# result = [w for w in engdict if re.search('^[ghi][mno][jlk][def]$', w)] #첫번째는 ghi중 하나, 두번째는 mno 중 하나, 세번째는 jlk중 하나, 네번째는 def 중 하나로 구성된 말을 찾아라
# result = [w for w in engdict if re.search('^[ah]+$', w)][:10]    #a혹은 h의 문자가 하나 이상 나오는 모든 단어를 찾아라(9번째까지)
print(result[:10])
```

```python
nltk.download('treebank')
wsj = nltk.corpus.treebank.words()

#ed 또는 ing로 끝나는 말들의 리스트를 만들어라
result = [w for w in wsj if re.search('(ed|ing)$', w)]
# result = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)] #숫자가 하나 이상 나오고 .(주의\.으로 써야함)이 나오고 숫자가 나오는 수를 찾아라 = 소숫점을 찾아라
# result = [w for w in wsj if re.search('^[A-Z]+\$$', w)] #대문자 알파벳이 하나 이상 나오고 $(\$=문자로서의 $)가 나오고 끝나는 문자를 찾아라
# result = [w for w in wsj if re.search('^[0-9]{4}$', w)]  #숫자가 정확하게 4개 나오는 수를 찾아라
# result = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)] #숫자가 나오고 -가 나오고 소문자가 3개 이상 5개 이하 나오는 수를 찾아라
# result = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)] #소문자가 5개 이상 나오고 -가 나오고 숫자가 2개 이상 3개 이상 나오고 -가 나오고 숫자가 6개 이하 나오는 수를 찾아라

result = sorted(set(result)) #오름차순 정렬해주는 기능
print(result[:10])
```

### Extract information (pos tag, named entity)

= 자연어에서 information을 뽑아내는 기능

* pos tag = part of speech tagging = 품사 태깅
* named entity = 카테고리를 알려줌

```python
sent = "I am Jhon from America and would like to go to Starbuck"
words = nltk.word_tokenize(sent)
words    #tokenize한 결과
```

```python
'''
POS tag list:

CC	coordinating conjunction
CD	cardinal digit
DT	determiner
EX	existential there (like: "there is" ... think of it like "there exists")
FW	foreign word
IN	preposition/subordinating conjunction
JJ	adjective	'big'
JJR	adjective, comparative	'bigger'
JJS	adjective, superlative	'biggest'
LS	list marker	1)
MD	modal	could, will
NN	noun, singular 'desk'
NNS	noun plural	'desks'
NNP	proper noun, singular	'Harrison'
NNPS	proper noun, plural	'Americans'
PDT	predeterminer	'all the kids'
POS	possessive ending	parent's
PRP	personal pronoun	I, he, she
PRP$	possessive pronoun	my, his, hers
RB	adverb	very, silently,
RBR	adverb, comparative	better
RBS	adverb, superlative	best
RP	particle	give up
TO	to	go 'to' the store.
UH	interjection	errrrrrrrm
VB	verb, base form	take
VBD	verb, past tense	took
VBG	verb, gerund/present participle	taking
VBN	verb, past participle	taken
VBP	verb, sing. present, non-3d	take
VBZ	verb, 3rd person sing. present	takes
WDT	wh-determiner	which
WP	wh-pronoun	who, what
WP$	possessive wh-pronoun	whose
WRB	wh-abverb	where, when
'''
```

```python
#앞에서 pos_tag하기 위해 tokenize한 결과값을 words에 저장했었음

nltk.download('averaged_perceptron_tagger')
pos = nltk.pos_tag(words)
pos
```

```python
nltk.download('maxent_ne_chunker')
NE = nltk.ne_chunk(pos)        #name entity = 해당하는 카테고리 태그-common entity-를 출력
                                #중요 정보를 뽑을 때 name entity가 중요
print(NE)
# common Entity types: ORGANIZATION, PERSON, LOCATION, DATE, TIME, MONEY, and GPE (geo-political entity)
```

*참고

- context의 정보에는 두 가지가 있음 = 문장에서 정보를 뽑아내는 방법

1) Named entity 정보

2) intention 정보

ex. 스타벅스에 가고싶다 => 스타벅스:named entity / 가고싶다:intention

ex. sk 미래관에서 기말고사 준비를 해야지 => sk 미래관, 기말고사 준비 : named entity / 공부를 할것이다 : intention

### Wordcloud

= 단어 등장 빈도 그리기

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

text = nltk.corpus.gutenberg.raw('bible-kjv.txt')

wc = WordCloud().generate(text)
plt.imshow(wc)  #여기서 만들어지는 그림은 frequency 기반
```

```python
stopwords = set(STOPWORDS)    #자주 등장하는 단어 = stopwords를 빼고 계산하는 법 => set의 형태로 넣어줘야 함!
stopwords.add('unto')
wc = WordCloud(stopwords = stopwords).generate(text) 
plt.imshow(wc)
```
